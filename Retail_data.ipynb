```python
df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Workspace/Users/vinilanair11@gmail.com/etl_repo/data/product_sales_dataset_final.csv")
display(df)
```

```python
from pyspark.sql.functions import sum

# Remove spaces from all column names
df = df.toDF(*[col.strip() for col in df.columns])

country_revenue = df.groupBy("Country").agg(sum("Revenue").alias("Total_Revenue"))
top_country = country_revenue.orderBy("Total_Revenue", ascending=False).limit(1)
display(top_country)
```
