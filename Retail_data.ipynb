```python
from pyspark.sql.functions import sum

# Remove spaces from all column names
df = df.toDF(*[col.strip() for col in df.columns])

country_revenue = df.groupBy("Country").agg(sum("Revenue").alias("Total_Revenue"))
top_country = country_revenue.orderBy("Total_Revenue", ascending=False).limit(1)
display(top_country)
```

```python
top_products = df.groupBy("Product_Name").agg(sum("Revenue").alias("Total_Revenue")).orderBy("Total_Revenue", ascending=False).limit(5)
display(top_products)
```

```python
from pyspark.sql.functions import sum

top_customer = df.groupBy("Customer_Name").agg(sum("Quantity").alias("Total_Products")).orderBy("Total_Products", ascending=False).limit(3)
display(top_customer)
```

```python
from pyspark.sql.functions import sum

# Calculate total quantity sold for each product in each country (region)
product_region_popularity = df.groupBy("Country", "Product_Name").agg(sum("Quantity").alias("Total_Quantity"))

# Find the most popular product in each region
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window_spec = Window.partitionBy("Country").orderBy(product_region_popularity["Total_Quantity"].desc())
popular_products_by_region = product_region_popularity.withColumn("rank", row_number().over(window_spec)).filter("rank = 1").drop("rank")

display(popular_products_by_region)
```

```python
from pyspark.sql import functions as F
import pyspark.sql.functions as F

monthly_revenue = (
    df.withColumn("month", F.date_trunc("month", F.col("Order_Date")))
      .groupBy("month")
      .agg(
          F.sum("Revenue").alias("total_revenue")
      )
      .orderBy("month")
)

monthly_revenue.show()
```
